```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

To get user agent : website > inspect > network tab > refresh the site > select any of the requests > go to header > scroll down to request header > copy user agent

```python
name_list=[]
rating_list=[]
company_type=[]
number_of_employees=[]
available_type=[]
company_age=[]
location=[]

review_count=[]
salary_count=[]
interview_count=[]
job_count=[]
benifit_count=[]
photo_count=[]

for k in range(1,500):
    header={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(f"https://www.ambitionbox.com/list-of-companies?page={k}",headers=header).text
    soup=BeautifulSoup(webpage,"lxml")
    company=soup.find_all("div",class_="companyCardWrapper__metaInformation")
    company_details=soup.find_all("div",class_="companyCardWrapper__tertiaryInformation")

    for i , j  in zip(company,company_details):
        name=i.find("h2").text.strip()
        rating=i.find("span",class_="companyCardWrapper__companyRatingValue").text.strip()
        all_types=i.find("span",class_="companyCardWrapper__interLinking").text.strip()
        all_list=all_types.split(" | ")

        data=j.find_all("span",class_="companyCardWrapper__ActionCount")


        if len(all_list)==5:
            name_list.append(name)
            rating_list.append(rating)
            company_type.append(all_list[0])
            number_of_employees.append(all_list[1])
            available_type.append(all_list[2])
            company_age.append(all_list[3])
            location.append(all_list[4])

            review_count.append(data[0].text)
            salary_count.append(data[1].text)
            interview_count.append(data[2].text)
            job_count.append(data[3].text)
            benifit_count.append(data[4].text)
            photo_count.append(data[5].text)
        else:
            continue
    print(k)
```

```python
dic={"name":name_list,"company_type":company_type,"rating":rating_list,"number_of_employees":number_of_employees,"available_type":available_type,"company_age":company_age,"location":location,"review_count":review_count,"salary_count":salary_count,"interview_count":interview_count,"job_count":job_count,"benifit_count":benifit_count,"photo_count":photo_count}
```

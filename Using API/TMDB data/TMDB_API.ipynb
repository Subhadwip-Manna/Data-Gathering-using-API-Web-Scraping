```python
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
final_df=pd.DataFrame()
```

```python
for i in range(1,570):
    response=requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=2225e849271081315aa6b980eb74d1ee&page={i}")
    data=pd.DataFrame(response.json()["results"])[["id","title","original_language","overview","popularity","release_date","vote_average","vote_count"]]
    final_df=final_df.append(data,ignore_index=True)
    print(i)
    
```
